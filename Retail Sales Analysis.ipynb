```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("supermarket_sales.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
# Data quality check
df.isnull().sum()
```

```python
# Total and average sales
total_sales = df['Total'].sum()
avg_transaction = df['Total'].mean()
print("Total Sales:", total_sales)
print("Average Transaction:", avg_transaction)
```

```python
# Daily sales trend
daily_sales = df.groupby('Date')['Total'].sum()
daily_sales.plot(title="Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Top-selling products
top_products = df.groupby('Product')['Quantity'].sum().sort_values()
top_products.plot(kind='barh', title="Top Selling Products")
plt.show()
```
